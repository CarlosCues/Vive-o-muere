```python
from extract_landmarks import extract_landmarks
from os import listdir
import cv2
import pandas as pd
import re
import numpy as np
from os import remove
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPool2D, BatchNormalization, Dropout, Flatten, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score
import joblib  
```

```python
basepath=[f'train/{imagen}' for imagen in listdir('train') ]
```

```python
landmarks_train=map(extract_landmarks,basepath)
landmarks_train=list(landmarks_train)
```

```python
dici = {basepath:landmarks_train for (basepath,landmarks_train) in zip(basepath,landmarks_train)}
```

```python
for k, v in dici.copy().items():
    if type(v)!=np.ndarray:
        del dici[k]
```

```python
lista= []
for x in landmarks_train:
    if type(x)==np.ndarray:
        lista.append(x)
        
        
```

```python
df=pd.DataFrame(dici)
df=pd.DataFrame(dici).T
df=df.reset_index()
```

```python
df
```

```python
for x in df['index']:
    if 'down' in x:
        df['index']=df['index'].replace(x,'down')
    if 'up' in x:
        df['index']=df['index'].replace(x,'up')
```

```python
df
```

```python
enc = LabelEncoder()
y=df['index']
y_=enc.fit_transform(y)
```

```python
X=df.drop(columns='index')
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y_)
```

```python
pipe_SVC = Pipeline([('scaler', StandardScaler()), ('svc', SVC())])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
modsave=joblib.dump(pipe_SVC, 'modelo_entrenado_pipe_SVC.pkl') # Guardo el modelo.
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('RF', RandomForestClassifier(n_estimators=1200))])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```
